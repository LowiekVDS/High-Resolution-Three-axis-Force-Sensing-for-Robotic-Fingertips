# STAGE I

```python
taxels_data = []
angle = -3.67 # in degrees

for i in range(ARRAY_SIZE):
  
  # Load the data
  taxel_data = cf.prepare_data_for_fitting(f"{TAXEL_DATA_FOLDER}{i}", ARRAY_SIZE, SENSOR_LAG, False)
  
  # Rotate the data labels by angle
  F_x = taxel_data['F_x'] * np.cos(np.deg2rad(angle)) - taxel_data['F_y'] * np.sin(np.deg2rad(angle))
  F_y = taxel_data['F_x'] * np.sin(np.deg2rad(angle)) + taxel_data['F_y'] * np.cos(np.deg2rad(angle))
  
  taxel_data['F_x'] = F_x
  taxel_data['F_y'] = F_y
  
  taxels_data.append(taxel_data)
```

```python
models = []
MSEs = []

for i in range(ARRAY_SIZE):
    
    taxel_data = taxels_data[i]
    
    # Fit the 0.5Hz data to a third order polynomial model
    columns = [f'X{i}', f'Y{i}',f'Z{i}']

    X = taxel_data[columns][:int(len(taxel_data) / 2)]
    Y = taxel_data[['F_x', 'F_y', 'F_z']][:int(len(taxel_data) / 2)]
    
    print(f"\nFitting taxel {i}:")
    model = reg.create_regression_pipeline_and_fit(X, Y, True)
    model.taxel_index = i
    models.append(model)
    
    # MSE 
    Y_pred = model.predict(X)
    mse = reg.mean_squared_error(taxel_data[['F_x', 'F_y', 'F_z']][:int(len(taxel_data) / 2)], Y_pred)
    MSEs.append(mse)
```

```python
# Also make one big giant model --> phase I
class StageIModel:
  def __init__(self, component_models) -> None:
    
    self.component_models = component_models
  
  def predict(self, df):
    
    Ys = []
    for i in range(len(self.component_models)):
      model = self.component_models[i]
      X = df[[f'X{i}', f'Y{i}',f'Z{i}']]
      Y = model.predict(X)
      Ys.append(Y)
    
    return np.array(Ys)
  
stage_I_model = StageIModel(models)
```

```python
# These contributions must be calibrated out. So we try to fit a model that models the influences it has to neighbours. These models 
# are then used in a feedforward compensation model
neighbour_elements = []

def get_neighbors(x, y):
    neighbors = []
    for i in range(x - 1, x + 2):
        for j in range(y - 1, y + 2):
            neighbors.append((i, j))
    return neighbors

for i in range(8):
  for j in range(4):
    
    center_value = i * 4 + j
    
    neighbors = get_neighbors(i, j)
    neighbors_values = [n[0] * 4 + n[1] for n in neighbors]
    real_values = []
    
    for coord in neighbors:
      if coord[0] >= 0 and coord[1] >= 0 and coord[0] <= 7 and coord[1] <= 3:
        real_values.append(coord[0] * 4 + coord[1])
    
    # Remove taxel self
    real_values.remove(center_value)
    
    neighbour_elements.append(real_values)

print(neighbour_elements)
```

```python
stage_Isub_models = []

for i in range(32):
  stage_Isub_models.append([])

for i in range(32):
  
  print(f"Doing neighbours of taxel {i}")
  
  # So we take a taxel and use its raw inputs as X
  # then we create a model per neighbour

  # The input vector is just the taxel inputs as usual
  X = taxels_data[i][1000:75000][[f'X{i}', f'Y{i}', f'Z{i}']]

  neighbours = neighbour_elements[i]

  # We predict the forces to then use as labels
  forces = stage_I_model.predict(taxels_data[i][1000:75000])    
  
  for neighbour in neighbours:
    
    Y = forces[neighbour]
    
    print(f"Influence of taxel {i} to {neighbour}")
    
    model = reg.create_regression_pipeline_and_fit(X, Y, degree=3)
    model.input_taxel_id = i
    model.output_taxel_id = neighbour
    
    stage_Isub_models[neighbour].append(model)
```

```python
# We use calibration on taxel 1 as a test dataset
dataset = taxels_data[1][1000:75000]
orig_forces = stage_I_model.predict(dataset)

# Now calculate the corrections for ex. taxel 0
TAXEL_ID = 0
pf.plot_three_components(orig_forces[TAXEL_ID].T) # Should be zero...

correction_forces = []

for model in stage_Isub_models[TAXEL_ID]:
  
  X = dataset[[f'X{model.input_taxel_id}', f'Y{model.input_taxel_id}', f'Z{model.input_taxel_id}']]
  forces = model.predict(X)

  pf.plot_three_components(forces.T)
  
  correction_forces.append(forces)
  
# Now correct orig_forces
new_forces = orig_forces[TAXEL_ID]
for f in correction_forces:
  new_forces -= f

pf.plot_three_components(new_forces.T)
```

```python
# An alternative way is to calculate the inproduct between the expected signals and the real signals of the neighbours
# So for example to compensate taxel 0, we take the initial prediction of taxel 0, we then calculate the inproduct between the initial prediction of stage I
# and what such an activation of taxel 0 would cause and rescale based on that
stage_I2sub_models = []

# for i in range(32):
#   stage_I2sub_models.append([])

for i in range(32):
  
  print(f"Doing neighbours of taxel {i}")
  
  # So we take a taxel and use its raw inputs as X
  # then we create a model per neighbour

  # The input vector is just the taxel inputs as usual
  X = taxels_data[i][1000:75000][[f'X{i}', f'Y{i}', f'Z{i}']]

  # We predict the forces to then use as labels
  forces = stage_I_model.predict(taxels_data[i][1000:75000])    
  
  Y = np.zeros((74000, 3*32))
  
  for j in range(32):
    Y[:, 3 * j : (j+1) * 3] = forces[j, :, :]
  
  model = reg.create_regression_pipeline_and_fit(X, Y, degree=3)
  
  # for j in range(32):
    
  #   Y = forces[j]
    
  #   print(f"Influence of taxel {i} to {j}")
    
  #   model = reg.create_regression_pipeline_and_fit(X, Y, degree=3)
  #   model.input_taxel_id = i
  #   model.output_taxel_id = j
    
  stage_I2sub_models.append(model)
```

```python
# Now we want to compensate taxel 0 based on the dataset of calibration taxel 1

# We use calibration on taxel 1 and 8 as a test dataset
dataset1 = taxels_data[1][1000:75000]
orig_forces1 = stage_I_model.predict(dataset1)
dataset2 = taxels_data[17][1000:75000]
orig_forces2 = stage_I_model.predict(dataset2)

orig_forces =  orig_forces2 + orig_forces1

# orig_forces has shape (taxel_nrs, ... , nr_axes)

all_rms = []

for k in range(32):

  TAXEL_ID = k

  # So per model, calculate the expected response should the original data have been by activating taxel TAXEL_ID
  expected_forces1 = stage_I2sub_models[TAXEL_ID].predict(dataset1[[f'X{TAXEL_ID}', f'Y{TAXEL_ID}', f'Z{TAXEL_ID}']])
  expected_forces2 = stage_I2sub_models[TAXEL_ID].predict(dataset2[[f'X{TAXEL_ID}', f'Y{TAXEL_ID}', f'Z{TAXEL_ID}']])

  # Expected forces shape is (... , taxel_nrs * nr_axes)
  expected_forces = expected_forces2 + expected_forces1
 
  print(orig_forces.shape, expected_forces.shape)
  
  error = 0
  for j in range(32):
    for k in range(3):
        
        error += np.sum((orig_forces[j, :, k] - expected_forces[:, k + j * 3]) ** 2)
        
  all_rms.append(error)
  
all_rms_rec = np.reciprocal(all_rms)

all_rms_rec /= np.max(all_rms_rec)

print(all_rms_rec)

for i in [0, 1, 4, 5, 17]:
  new_forces = orig_forces[i]
  new_forces[:, 0] *= all_rms_rec[i]
  new_forces[:, 1] *= all_rms_rec[i]
  new_forces[:, 2] *= all_rms_rec[i]

  pf.plot_three_components(new_forces.T)
        
```

```python
# Split the data
def create_dataset(dataset, nr_steps):
    dataX, dataY = [], []
    for i in range(dataset.shape[0] - nr_steps - 1):
        a = dataset[i:(i + nr_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + nr_steps, 1])
    return np.array(dataX), np.array(dataY)

def create_dataset_single(dataset, nr_steps):
    dataX = []
    for i in range(dataset.shape[0] - nr_steps - 1):
        a = dataset[i:(i + nr_steps)]
        dataX.append(a)
    return np.array(dataX)

n_steps = 500
features = 1

# split into samples
X_train, y_train = create_dataset(train, n_steps)
X_test, y_test = create_dataset(test, n_steps)

X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
from keras.losses import MeanSquaredError
from keras.optimizers import Adam

# Define LSTM model
model = Sequential()

# model.add(LSTM(64, return_sequences=True))  
model.add(LSTM(16))  
model.add(Dense(1))
# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.compile(optimizer=Adam(0.001), loss=MeanSquaredError(), metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=30, batch_size=16)
```
