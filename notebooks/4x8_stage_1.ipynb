# Setup

First we setup the creation by loading in all data

```python
import numpy as np

from tools import common_functions as cf
from tools import regression as reg

# Folder prefix with all the taxel calibration data
TAXEL_DATA_FOLDER = f"table_calibration/taxel"

# Parameters
SENSOR_LAG = 80 # ms
ARRAY_SIZE = 32
ALTERNATIVE_CALIBRATION = False

data_columns = [f'X{i}' for i in range(ARRAY_SIZE)] + [f'Y{i}' for i in range(ARRAY_SIZE)] + [f'Z{i}' for i in range(ARRAY_SIZE)]
```

# STAGE I

```python
taxels_data = []

for i in range(ARRAY_SIZE):
  
  if ALTERNATIVE_CALIBRATION:
    taxel_data = cf.prepare_data_for_fitting(f"{TAXEL_DATA_FOLDER}{i}", ARRAY_SIZE, SENSOR_LAG, rotation=90 - 1.82, faulty=False, cut_in_half=False)
    taxel_data['F_x'] *= -1 # Setup error! Used left handed coordinate frame :(
  else:  
    taxel_data = cf.prepare_data_for_fitting(f"{TAXEL_DATA_FOLDER}{i}", ARRAY_SIZE, SENSOR_LAG, rotation=90 + 1.82, faulty=False, cut_in_half=True)
  
  taxels_data.append(taxel_data)
```

```python
lin_models = []
models = []
MSEs = []
scores = []

for degree in [1,3,5]:
    
    MSEtmp = []
    scoretmp = [] 
     
    for i in range(ARRAY_SIZE):
        
        taxel_data = taxels_data[i]
        
        # Fit the 0.5Hz data to a third order polynomial model
        columns = [f'X{i}', f'Y{i}',f'Z{i}']

        X = taxel_data[columns]
        Y = taxel_data[['F_x', 'F_y', 'F_z']]
        
        # X = taxel_data[columns][taxel_data['F_z'] > -6]
        # Y = taxel_data[['F_x', 'F_y', 'F_z']][taxel_data['F_z'] > -6]
        
        # X = taxel_data[columns][np.abs(taxel_data['F_x']) < 2.5][np.abs(taxel_data['F_y']) < 2.5]
        # Y = taxel_data[['F_x', 'F_y', 'F_z']][np.abs(taxel_data['F_x']) < 2.5][np.abs(taxel_data['F_y']) < 2.5]
        
        print(f"\nFitting taxel {i}:")
        model, _, X_test, _, Y_test = reg.create_regression_pipeline_and_fit(X, Y, False, degree=degree)
        model.taxel_index = i
        
        if degree == 3:
            models.append(model)
            
        if degree == 1:
            lin_models.append(model)
        
        # MSE 
        Y_pred = model.predict(X_test)
        N, _ = Y_pred.shape
        
        mse_custom = np.sum((Y_test - Y_pred) ** 2) / N    

        max = np.max(Y_pred, axis=0)
        min = np.min(Y_pred, axis=0)
        
        diff = max - min
        
        print(diff)
        
        for i in range(3):
            mse_custom[i] = mse_custom[i] / diff[i]
            
        mse_custom = np.sum(mse_custom) / 3
        
        MSEtmp.append(mse_custom)
        scoretmp.append(model.score(X_test, Y_test))
        
        print(f"MSE: {mse_custom}")
        print(f"Score: {model.score(X_test, Y_test)}")
        
    MSEs.append(MSEtmp)
    scores.append(scoretmp)
```

```python
# Construct the latex table body
table_body = ""

for i in range(ARRAY_SIZE):
    table_body += f"Taxel {i} "
    
    for j in range(3):
        table_body += f"& \multicolumn{{1}}{{l|}}{{{scores[j][i]:.5f}}} & {MSEs[j][i]:.5f} "
    
    table_body += "\\\\ \hline \n"

print(table_body)
```

```python
for j in range(3):
  
  # Calc mean score and mse
  mean_score = np.mean(scores[j])
  mean_mse = np.mean(MSEs[j])
  
  # Calc std score and mse
  std_score = np.std(scores[j])
  std_mse = np.std(MSEs[j])    
  
  print(f"Degree {2*j + 1}: Mean score: {mean_score:.5f} +- {std_score:.5f}, Mean MSE: {mean_mse:.5f} +- {std_mse:.5f}")
```

```python
# Also make one big giant model --> phase I
class StageIModel:
  def __init__(self, component_models) -> None:
    
    self.component_models = component_models
  
  def predict(self, df):
    
    Ys = []
    for i in range(len(self.component_models)):
      model = self.component_models[i]
      X = df[[f'X{i}', f'Y{i}',f'Z{i}']]
      Y = model.predict(X)
      Ys.append(Y)
    
    return np.array(Ys)
  
  def predict_single(self, X):
    
    # X: shape is (N, 3)
    
    Ys = []
    for i in range(len(self.component_models)):
      model = self.component_models[i]
      Y = model.predict(X[i].reshape(1,-1))
      Ys.append(Y)
    
    return np.array(Ys)
  
stage_I_model = StageIModel(models)
reg.save_taxel_models(stage_I_model, f"4x8", f"stageI")
```
